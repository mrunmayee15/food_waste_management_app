```python
# Importing required libraries

%pip install pandas numpy
%pip install matplotlib
import pandas as pd
print(pd.__version__)
import numpy as np
import matplotlib.pyplot as plt
```

```python
#loading dataset
file_path = "claims_data.csv"
df = pd.read_csv(file_path)
```

```python
#dataset information

print("First 5 rows :")
print(df.head())
print(f"Total Rows: {df.shape[0]} and Total Columns: {df.shape[1]}")

print("\nData types of columns")
print(df.dtypes)
```

```python
#missing values

print("\nMissing values per column :")
missing_counts = df.isna().sum()
print(missing_counts)
```

```python
#duplicate rows

print(f"Duplicate Rows Count : {df.duplicated().sum()}")
```

```python
#understanding variables

print("\nColumn Summary (Numeric type Only) :")

numeric_cols = df.select_dtypes(include=[np.number]).columns

if len(numeric_cols) > 0:
    print(df[numeric_cols].describe())
else:
    print("Dataset doesn't have any numeric type column")
```

```python
#handle datatypes
df['Claim_ID'] = df['Claim_ID'].astype('int64')
df['Food_ID'] = df['Food_ID'].astype('int64')
df['Receiver_ID'] = df['Receiver_ID'].astype('int64')

# Convert Status to category (recommended)
df['Status'] = df['Status'].astype('category')

# Convert Timestamp to datetime
df['Timestamp'] = pd.to_datetime(df['Timestamp'], errors='coerce')

# Verify changes
print(df.dtypes)
```

```python
#timestamp formatting

df['Timestamp'] = pd.to_datetime(
    df['Timestamp'],
    errors='coerce'
)

print(df['Timestamp'].dtype)
print(df['Timestamp'].isnull().sum())
```

```python
#create new file of cleaned data
df.to_csv("Claims_Data_Cleaned.csv", index=False)
```
